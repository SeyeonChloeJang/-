```python
def Calculator(x, y, operator):
  if operator == "+":
    result = x + y
  elif operator == "-":
    result = x - y
  elif operator == "*":
    result = x * y
  elif operator == "/":
    result = x / y
  elif operator == "%":
    result = x % y
  else :
    print ("다른 연산자를 입력하세요")
  return result
```

```python
Calculator(3,4,"*")
```

Problem 2

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
arr = np.random.randint(0,10,size=(10,10))
print(arr)
```

```python
value_counts = pd.Series(arr.ravel()).value_counts()

df = pd.DataFrame({'Number': value_counts.index, 'Frequency': value_counts.values})
print(df)
```

```python
plt.hist(arr.ravel(), bins=np.arange(-0.5, 10.5, 1), edgecolor='black', rwidth = 0.8)
plt.xticks(np.arange(10))
plt.xlabel('Number')
plt.ylabel('Frequency')
plt.title('Histogram of Integer Frequencies')
plt.show()
```
