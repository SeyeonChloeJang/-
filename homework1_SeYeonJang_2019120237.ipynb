```python
# 문제 해석 1: 학번 뒤 6자리는 120237이며, 1000번 반복하여 해당 숫자열이 나오는 개수의 평균값을 구함
# 주어진 문제가 하나의 '100만개 숫자가 있는 열'에 '몇 개의 특정 숫자열이 등장하는지' 알아보고자 하는 걸로 해석했음
# (문제: Generate 'one million random single digit numbers' and find out 'how many' sequences of numbers that matches exactly the last 6 digits of your student number.)
import random
count = 0
n = 1000
for i in range(n):
    random_string = ''.join(str(random.randint(0, 9)) for _ in range(1000000))
    count += random_string.count('120237')

print(f"120237이 1000번 시행 동안 나타난 횟수: {count}")
print(f"120237이 1번 시행 동안 나타나는 평균 횟수: {count/1000: .3f}")
```

```python
# 문제 해석 2: 길이가 백만인 랜덤 숫자열에 특정 6자리 숫자가 존재할 확률 구하기
import numpy as np
array_size = 1000000
trial = 1000

count = 0
for j in range(trial):
  array = np.random.randint(0,10,size = array_size)
  for i in range(array_size - 5):
    if array[i] == 1:
      if array [i+1] == 2:
        if array [i+2] == 0:
          if array [i+3] == 2:
            if array [i+4] == 3:
              if array [i+5] == 7:
                count += 1
                break

print(f"길이가 백만인 랜덤 숫자열에 특정 6자리 숫자가 존재할 확률: {count/trial}")
```

```python
# Scatterplot of house value
import pandas as pd
import matplotlib.pyplot as plt

housing = pd.read_csv("https://raw.githubusercontent.com/ageron/handson-ml/master/datasets/housing/housing.csv")
plt.figure(figsize=(15,15))
plt.plot('median_house_value',
         'median_income',
         data = housing, 
         linestyle = 'none', 
         marker = 'o', 
         markersize = 2,
         color = 'blue', 
         alpha = 1)
plt.title('Scatterplot of House Value', fontsize=20)
plt.xlabel('Median House Value', fontsize=14)
plt.ylabel('Median Income', fontsize=14)
plt.show()
```

```python
# 새로운 column 생성
df_new = pd.DataFrame(housing)
df_new['new_variable'] = df_new['total_bedrooms'] / df_new['households']
df_new
```

```python
# 구간 나누기 경우 1: 동일한 구간크기 10개로 나누기
cut = pd.cut(df_new['new_variable'], 10).value_counts()
cut
# 매우 작은 범위에 데이터가 몰려 있음
# [최솟값, 최댓값]을 동일 크기로 나눈 bin 안의 frequency를 세는 것은 무의미하다고 판단되어, quartile 단위로 나눌 것
```

```python
# 구간 나누기 경우 2: quartile 단위로 나누기
qcut = pd.qcut(df_new['new_variable'], q = [0, 0.25, 0.5, 0.75, 1]).value_counts()
qcut
# 하지만 그럼에도 불구, 아주 작은 두 번째 구간에 매우 몰려 있음
```

```python
# quartile에 따라 label 매기기
df_new['quartiles'] = pd.qcut(df_new['new_variable'], 4, labels = ['1', '2', '3', '4'])
print(df_new)
```

```python
# 각 그룹 별 새로운 변수의 평균, 빈도수 구하기
df_new = df_new[['new_variable', 'quartiles']]
df_new.groupby(['quartiles']).agg(['mean', 'count'])
```

```python
# Student 선택, Course class가 필요하여 추가함
class Course:
  def __init__(self, name, credits, grade):
    self.name = name
    self.credits = credits
    self.grade = grade

  def grade_scale(self):
    if self.grade == "A+":
      return 4.5
    elif self.grade == "A":
      return 4.0
    elif self.grade == "B+":
      return 3.5
    elif self.grade == "B":
      return 3.0
    elif self.grade == "C+":
      return 2.5
    elif self.grade == "C":
      return 2.0
    elif self.grade == "D+":
      return 1.5
    elif self.grade == "D":
      return 1.0
    else:
      return 0.0

  def points_times_credits(self):
    return self.grade_scale() * self.credits

class Student:
  def __init__(self, name, age, major):
    self.name = name
    self.age = age
    self.major = major
    self.courses = []

  def about_me(self):
    print("My name is", self.name, "and I am", self.age, "years old.")

  def register(self, course):
    self.courses.append(course)
  
  def calculate_gpa(self):
    total_points = 0
    total_credits = 0
    for course in self.courses:
      total_points += course.points_times_credits()
      total_credits += course.credits
    return total_points / total_credits

  def get_major(self):
    return self.major
```

```python
# Exmaple 1: 자기소개하기
h = Student("Harry", 19, "mathematics")
h.about_me()
```

```python
# Example 2: 전공 알아보기
s = Student("Chloe", 22, "Business Administration")
major = s.get_major()
print("Chloe majors in", major)
```

```python
# Example 3: 학점 계산하기
a = Student("Alisa", 22, "Political Science")
course1 = Course("PSYC101", 3, "A+")
course2 = Course("PSIS203", 3, "B")
course3 = Course("MATH104", 1.5, "B+")

a.register(course1)
a.register(course2)
a.register(course3)

print("Alisa's gpa: ", a.calculate_gpa())
```
